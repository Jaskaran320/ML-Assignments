## Part 5

```python
data_and = utils.xao_dataset.And()
data_or = utils.xao_dataset.Or()
data_xor = utils.xao_dataset.Xor()
perceptron_and = utils.perceptron(data_and, alpha=0.01, epochs=50)
perceptron_or = utils.perceptron(data_or, alpha=0.01, epochs=50)
perceptron_xor = utils.perceptron(data_xor, alpha=0.01, epochs=50)
```

### Learnable bias

```python
w_and, b_and, miss_and = perceptron_and.train()
w_and, b_and
perceptron_and.plot(w_and)
```

```python
w_or, b_or, miss_or = perceptron_or.train()
w_or, b_or
perceptron_or.plot(w_or)
```

```python
w_xor, b_xor, miss_xor = perceptron_xor.train()
w_xor, b_xor
perceptron_xor.plot(w_xor)
```

### Fixed bias

```python
w_and_0, b_and_0, miss_and_0 = perceptron_and.train(bias_0=True)
w_and_0, b_and_0, miss_and_0
perceptron_and.plot(w_and_0)
```

```python
w_or_0, b_or_0, miss_or_0 = perceptron_or.train(bias_0=True)
w_or_0, b_or_0
perceptron_or.plot(w_or_0)
```

```python
w_xor_0, b_xor_0, miss_xor_0 = perceptron_xor.train(bias_0=True)
w_xor_0, b_xor_0
perceptron_xor.plot(w_xor_0)
```
